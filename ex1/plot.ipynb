# Memory Access Stride Analysis

Comparing performance with -O0 vs -O2 optimization

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Data

Load the performance data from both optimization levels.

```python
# Load data
df_o0 = pd.read_csv('stride_o0.csv')
df_o2 = pd.read_csv('stride_o2.csv')

print("Data loaded successfully!")
print(f"O0 shape: {df_o0.shape}")
print(f"O2 shape: {df_o2.shape}")
```

## Execution Time Comparison

```python
plt.figure(figsize=(10, 5))
plt.plot(df_o0['stride'], df_o0['time (msec)'], 'o-', label='O0 (No optimization)', linewidth=2)
plt.plot(df_o2['stride'], df_o2['time (msec)'], 's-', label='O2 (Optimized)', linewidth=2)
plt.xlabel('Stride')
plt.ylabel('Time (msec)')
plt.title('Execution Time vs Stride')
plt.legend()
plt.grid(True)
plt.show()
```

## Memory Bandwidth Comparison

```python
plt.figure(figsize=(10, 5))
plt.plot(df_o0['stride'], df_o0['rate (MB/s)'], 'o-', label='O0 (No optimization)', linewidth=2)
plt.plot(df_o2['stride'], df_o2['rate (MB/s)'], 's-', label='O2 (Optimized)', linewidth=2)
plt.xlabel('Stride')
plt.ylabel('Bandwidth (MB/s)')
plt.title('Memory Bandwidth vs Stride')
plt.legend()
plt.grid(True)
plt.show()
```

## Key Observations

- **O2 is faster**: Compiler optimizations (loop unrolling, vectorization) significantly improve performance
- **Stride = 1 is best**: Sequential memory access exploits cache efficiently
- **Performance degrades with stride**: Larger strides waste cache lines and cause more cache misses
- **Bandwidth drops**: As stride increases, effective memory bandwidth decreases proportionally
