```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("blocks.csv")

# Make sure numeric
for c in ["block", "time_ms", "gflops", "bandwidth_GBps"]:
    df[c] = pd.to_numeric(df[c])

df = df.sort_values("block")
```

```python
# Plot time vs block
plt.figure()
plt.plot(df["block"], df["time_ms"], marker="o")
plt.xlabel("Block size B")
plt.ylabel("Time (ms)")
plt.title("Blocked MXM: Time vs Block size")
plt.show()
```

```python
# Plot GFLOP/s vs block
plt.figure()
plt.plot(df["block"], df["gflops"], marker="o")
plt.xlabel("Block size B")
plt.ylabel("GFLOP/s")
plt.title("Blocked MXM: GFLOP/s vs Block size")
plt.show()
```

```python
# Plot bandwidth vs block (as reported)
plt.figure()
plt.plot(df["block"], df["bandwidth_GBps"], marker="o")
plt.xlabel("Block size B")
plt.ylabel("Bandwidth (GB/s)")
plt.title("Blocked MXM: Bandwidth vs Block size (reported)")
plt.show()
```

```python

best = df.loc[df["time_ms"].idxmin()]
print("Best block size by time:", int(best["block"]))
print(best)
```
